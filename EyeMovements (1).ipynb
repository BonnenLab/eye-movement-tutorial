Here is the code that produced the graph above. 

```python
import matplotlib.pyplot as plt # library for plotting
import numpy as np #library for basic math
import pandas as pd  #libary for database
 
url_Saccade = "https://raw.githubusercontent.com/BonnenLab/eye-movement-tutorial/main/Saccade_gaze_positions.csv"
df_Saccade = pd.read_csv(url_Saccade) #importing url into a database
 
time = df_Saccade.gaze_timestamp #naming a column 
normposX = df_Saccade.norm_pos_x 
normposY = df_Saccade.norm_pos_y

fig, axes = plt.subplots(2) # Splitting the page into two graphs
fig.set_figheight(6) #setting the size of the image
fig.set_figwidth(9)

## top graph ## 
axes[0].plot(time-time[0], normposX, color='r') # ploting  x line
axes[0].plot(time-time[0], normposY, color='g') # ploting y line
axes[0].set_xlabel('time') #labeling
axes[0].set_ylabel('positions')
axes[0].set_title('Saccade: x and y eye positions over time')
axes[0].legend(['X position', 'Y position']) #creating a key 

## bottom graph ## 
axes[1].scatter(normposX, normposY, color='g', s=15) # ploting the dots
axes[1].plot(normposX, normposY) # ploting the line
axes[1].set_xlabel('X eye postion')
axes[1].set_ylabel('Y eye position')
axes[1].set_title('X vs. Y positions')
fig.tight_layout(pad =.80)
plt.show()
```

    - Effect on Saccade: Increased chance of saccade dysmetria, reduced accuracy. "Decreased saccade amplitude and duration indicate spatially clustered fixations that attenuate visual exploration and emphasize endogenous over exogenous attention. We propose altered pontocerebellar motor modulation as underlying mechanism that contributes to atypical (oculo-)motor coordination and attention function in ASD." (Bast et. al)
    - NO effect on Smooth pursuit : "In addition to the finding of anticipatory smooth eye movements in both groups, the statistical analyses also verified that there were no significant differences between anticipatory eye velocities in the ASD and typical groups. This supports the central finding of this study, namely, that anticipatory smooth eye movements in high-functioning ASD are not different from those in typical populations." (Aitkin, et. al) 
    - NO effect on VOR: "The findings of this and prior studies reviewed above provide substantial support for the functional integrity of localized brainstem-cerebellar circuitry in autism, suggesting that histopathological abnormalities may not be associated with intrinsic dysfunction of these structures. Rather, the evidence suggests that these structural alterations more likely reflect altered development and maturation of cortical systems." (Furman et. al)
